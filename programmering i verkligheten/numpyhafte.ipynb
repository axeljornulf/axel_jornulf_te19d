```python
#Uppgift 3
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,10)

y = 3*x + 1

figure = plt.figure()
axle = figure.add_subplot(1,1,1)
axle.set_title('Graf till f(x) = 3x + 1')
axle.set_xlabel('x', loc='right')
axle.set_ylabel('y', loc='top')
axle.spines['left'].set_position('center')
axle.spines['bottom'].set_position('center')
axle.spines['right'].set_color('none')
axle.spines['top'].set_color('none')

plt.plot(x,y, 'r')

plt.show()
```

```python
#Uppgift 4
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2,2)

y = (x**2)-1

figure = plt.figure()
axle = figure.add_subplot(1,1,1)
axle.set_title('Graf till f(x) = x^2 - 1')
axle.set_xlabel('x', loc='right')
axle.set_ylabel('y', loc='top')
axle.spines['left'].set_position('center')
axle.spines['bottom'].set_position('center')
axle.spines['right'].set_color('none')
axle.spines['top'].set_color('none')

plt.plot(x,y, 'b')

plt.show()
```

```python
#Uppgift 5
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0,6)
ht = -4*t**2 + 24*t + 1

figure = plt.figure()
axle = figure.subplots()
axle.set_title('h(t) = -4t² + 24t + 1')
axle.set_xlabel('t sekunder', loc='right')
axle.set_ylabel('h(t) meter', loc='top')
axle.spines['right'].set_color('none')
axle.spines['top'].set_color('none')

plt.plot(t,ht, 'g')

plt.show()

print("Vid 3 sekunder, når raketen sin högsta höjd: 35 meter.")
```
