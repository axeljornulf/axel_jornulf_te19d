```python
#Uppgift 1 c)

import pandas as pd
import matplotlib.pyplot as plt


null = []

df = pd.read_csv('sysselsattning.csv', header=1)

ärArbetande = df["2019"].iloc[0:12].sum()
ejArbetande = df["2019"].iloc[12:24].sum()

null.append(ärArbetande)
null.append(ejArbetande)

labels = 'Är förvärvsarbetande', 'Ej förvärvsarbetande'

plt.pie(null, labels=labels, autopct='%1.1f%%')
plt.title("Andel som är förvärvsarbetande och ej förvärvsarbetande")
plt.axis('equal')
plt.show()
```

```python
#Uppgift 2 e)

import pandas as pd
import matplotlib.pyplot as plt

arbete = []
labels = "Män som \nförvärvsarbetar", "Män som \nej förvärvsarbetar", "Kvinnor som \nförvärvsarbetar", "Kvinnor som \nej förvärvsarbetar"

df = pd.read_csv('sysselsattning.csv', header=1)

mänJobbar = df.query('sysselsättning == "förvärvsarbetande" and kön == "män"')['2019'].sum()
mänEjJobbar = df.query('sysselsättning == "ej förvärvsarbetande" and kön == "män"')['2019'].sum()
kvinnorJobbar = df.query('sysselsättning == "förvärvsarbetande" and kön == "kvinnor"')['2019'].sum()
kvinnorEjJobbar = df.query('sysselsättning == "ej förvärvsarbetande" and kön == "kvinnor"')['2019'].sum()

arbete.append(mänJobbar)
arbete.append(mänEjJobbar)
arbete.append(kvinnorJobbar)
arbete.append(kvinnorEjJobbar)

plt.bar(labels, arbete, width=.4, color = ["blue", "green", "red", "pink"])
plt.title("Sysselsättning för män och kvinnor mellan 16-44 år gammal")
plt.show()
```
