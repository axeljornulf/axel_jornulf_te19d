```python
#Uppgift 1 c)

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

null = []

df = pd.read_csv('sysselsattning.csv', header=1)

ärArbetande = df["2019"].iloc[0:12].sum()
ejArbetande = df["2019"].iloc[12:24].sum()

null.append(ärArbetande)
null.append(ejArbetande)

labels = 'Är förvärvsarbetande', 'Ej förvärvsarbetande'

plt.pie(null, labels=labels, autopct='%1.1f%%')
plt.title("Andel som är förvärvsarbetande och ej förvärvsarbetande")
plt.axis('equal')
plt.show()
```
